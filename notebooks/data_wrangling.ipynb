```python
import pandas as pd

# Load the spotify data file into a DataFrame
tracks_df = pd.read_csv('../datasets/spotify_data.csv')

# Display the first few rows 
tracks_df.head()
```

```python
# Get a list of artists
artists = tracks_df['artist_name'].unique().tolist()

# Get the number of unique artists
len(artists)
```

```python
# Load artists file
artists_df = pd.read_csv('../datasets/artists.csv')

# Get artists with genres containing "pop"
pop_artists = artists_df[artists_df['genres'].str.contains('pop')]
pop_artists.head()
```

```python
# Display tracks for the artist "Keane"  
tracks_df.loc[tracks_df['artist_name'] == 'Keane']
```

```python
# For tracks with the genre "piano", change the genre to "pop" if the artist is a pop artist
piano_mask = tracks_df['genre'] == 'piano'
piano_artists = tracks_df.loc[piano_mask, 'artist_name']
tracks_df.loc[piano_mask, 'genre'] = piano_artists.apply(
    lambda artist: 'pop' if artist in pop_artists['name'].values else 'piano'
)

# Display tracks for the artist "Keane"  
tracks_df.loc[tracks_df['artist_name'] == 'Keane']
```

```python
# Get a list of genres
genres = tracks_df['genre'].unique()

# Output the list
genres
```

```python
# Map genres to categories
genre_to_category = {'classical': 'Classical', 'new-age': 'Classical', 'opera': 'Classical', 'piano': 'Classical', 'country': 'Country', 'forro': 'Country', 'sertanejo': 'Country', 'ambient': 'Electronic', 'breakbeat': 'Electronic', 'chicago-house': 'Electronic', 'chill': 'Electronic', 'club': 'Electronic', 'dance': 'Electronic', 'deep-house': 'Electronic', 'detroit-techno': 'Electronic', 'drum-and-bass': 'Electronic', 'dub': 'Electronic', 'dubstep': 'Electronic', 'edm': 'Electronic', 'electro': 'Electronic', 'electronic': 'Electronic', 'garage': 'Electronic', 'hardstyle': 'Electronic', 'house': 'Electronic', 'minimal-techno': 'Electronic', 'progressive-house': 'Electronic', 'sad': 'Electronic', 'sleep': 'Electronic', 'techno': 'Electronic', 'trance': 'Electronic', 'trip-hop': 'Electronic', 'acoustic': 'Folk', 'folk': 'Folk', 'singer-songwriter': 'Folk', 'songwriter': 'Folk', 'disco': 'Groove', 'funk': 'Groove', 'groove': 'Groove', 'soul': 'Groove', 'gospel': 'Gospel', 'afrobeat': 'Hip-Hop', 'dancehall': 'Hip-Hop', 'hip-hop': 'Hip-Hop', 'blues': 'Jazz', 'jazz': 'Jazz', 'black-metal': 'Metal', 'death-metal': 'Metal', 'grindcore': 'Metal', 'hardcore': 'Metal', 'hard-rock': 'Metal', 'heavy-metal': 'Metal', 'industrial': 'Metal', 'metal': 'Metal', 'metalcore': 'Metal', 'indie-pop': 'Pop', 'party': 'Pop', 'pop': 'Pop', 'pop-film': 'Pop', 'romance': 'Pop', 'show-tunes': 'Pop', 'alt-rock': 'Rock', 'emo': 'Rock', 'goth': 'Rock', 'guitar': 'Rock', 'power-pop': 'Rock', 'psych-rock': 'Rock', 'punk': 'Rock', 'punk-rock': 'Rock', 'rock': 'Rock', 'rock-n-roll': 'Rock', 'ska': 'Rock', 'cantopop': 'World', 'french': 'World', 'german': 'World', 'indian': 'World', 'k-pop': 'World', 'salsa': 'World', 'samba': 'World', 'spanish': 'World', 'swedish': 'World', 'tango': 'World'}

# Add new column with mapped genres
tracks_df['new_genre'] = tracks_df['genre'].map(genre_to_category)

# Remove rows with missing values
tracks_df = tracks_df.dropna(subset=['new_genre'])

# Display first few rows
tracks_df.head()
```

```python
# Save the DataFrame to a new CSV file
columns_to_save = ["track_id", "track_name", "artist_name", "new_genre", "valence", "energy", "popularity", "instrumentalness"]
tracks_df[columns_to_save].to_csv('../datasets/track_data.csv', index=False)
```
